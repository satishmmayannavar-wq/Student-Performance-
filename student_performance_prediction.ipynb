```python
# ---------------------------------------
# 1. Install Required Libraries
# ---------------------------------------
!pip install -U transformers torch pandas
# ---------------------------------------
# 2. Import Libraries
# ---------------------------------------
from transformers import pipeline
import pandas as pd
# ---------------------------------------
# 3. Load Model
# ---------------------------------------
print("Loading model...")
generator = pipeline(
    "text-generation",
    model="google/flan-t5-base"
)
# ---------------------------------------
# 4. Dataset
# ---------------------------------------
texts = [
    "the movie was fantastic and very entertaining",
    "I hated the service at the restaurant",
    "The product quality is amazing",
    "the delivery was very slow and disappointing",
    "the lecture was informative and useful"
]
true_labels = ["Positive", "Negative", "Positive", "Negative", "Positive"]
# ---------------------------------------
# 5. Zero-Shot Prompting
# ---------------------------------------
def zero_shot(text):
    prompt = f"Classify the sentiment (Positive or Negative): {text}"
    response = generator(prompt, max_length=30)
    return response[0]["generated_text"]
# ---------------------------------------
# 6. One-Shot Prompting
# ---------------------------------------
def one_shot(text):
    prompt = f"""
Example:
Text: I love this product
Sentiment: Positive

Now classify:
Text: {text}
Sentiment:
"""
    response = generator(prompt, max_length=30)
    return response[0]["generated_text"]

# ---------------------------------------
# 7. Few-Shot Prompting
# ---------------------------------------
def few_shot(text):
    prompt = f"""
Example 1:
Text: I love this product
Sentiment: Positive

Example 2:
Text: This is the worst service ever
Sentiment: Negative

Example 3:
Text: The food was delicious
Sentiment: Positive

Now classify:
Text: {text}
Sentiment:
"""
    response = generator(prompt, max_length=30)
    return response[0]["generated_text"]
# ---------------------------------------
# 8. Run Experiment
# ---------------------------------------
results = []
for i, text in enumerate(texts):

    zs = zero_shot(text)
    os = one_shot(text)
    fs = few_shot(text)

    results.append({
        "Text": text,
        "True_Label": true_labels[i],
        "Zero_Shot": zs,
        "One_Shot": os,
        "Few_Shot": fs
    })
# ---------------------------------------
# 9. Convert Results to DataFrame
# ---------------------------------------
df = pd.DataFrame(results)

# ---------------------------------------
# 10. Display Results
# ---------------------------------------
print("\nResults:\n")
print(df)




```
